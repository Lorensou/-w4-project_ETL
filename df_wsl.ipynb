```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Lorenzo\Desktop\df.wsl.csv")
df.head()
```

```python
df.drop('injuried_flag', axis=1, inplace=True) #Borro la ultima columna que no me interesa.
df.tail()
```

```python
# Ahora quiero borrar todos los datos que sean anteriores a 2022
mascara = df['year'] == 2022
df = df.loc[mascara]
df.shape
```

```python
f3 = df.groupby('athlete_name')['points_earned'].sum().reset_index()
df4 = df.groupby(['athlete_name']).agg({'points_earned': ['sum'], 'athlete_country': 'first', 'athlete_profile_url': 'first'}).reset_index()
df_wsl = df4
df_wsl
```

```python
df_wsl = df_wsl.dropna()
df_wsl
```

```python
df_wsl = df_wsl.reset_index(drop=True) 
df_wsl
```

```python

df_wsl = df_wsl.rename(columns={'athlete_name': 'athlete_name', 'points_earned': 'points_earned_2022', 'athlete_country': 'country', 'athlete_profile_url': 'profile_url'})
df_wsl
```

```python
df_wsl['points_earned_2022'].mean()
```

```python
df_wsl.columns = df_wsl.columns.droplevel(1) #Borro el nombre de columnas del nivel inferior
df_wsl
```

```python
df_wsl = df_wsl[df_wsl['points_earned_2022'] > 30000]
df_wsl = df_wsl.drop(df_wsl[df_wsl['athlete_name'].isin(['Miguel Pupo', 'Samuel Pupo'])].index)

df_wsl
```

```python
df_wsl = df_wsl.reset_index(drop=True)  #Ahora tengo los 11 surfistas con más puntos obtenidos en 2022 :).
df_wsl
```

```python
df_wsl = df_wsl.sort_values(by='points_earned_2022', ascending=False)
df_wsl
```

```python
df_wsl = df_wsl.reset_index(drop=True)
df_wsl
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import WebDriverException
import time
```

```python
#driver configuration
opciones=Options()

opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    # si True, no aperece la ventana (headless=no visible)
opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
opciones.add_argument('--incognito')
```

```python
PATH = ChromeDriverManager().install() #instalar driver en crhome
driver = webdriver.Chrome(PATH) #Para poder abrir una ventana en Crhome
```
